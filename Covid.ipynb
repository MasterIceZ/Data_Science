```python
import requests
import json
from matplotlib import pyplot as plt
```

```python
res = requests.get('https://raw.githubusercontent.com/wiki/djay/covidthailand/cases_briefings')
dat = json.loads(res.text)
try:
    print(type(dat))
    print(dat[0])
except:
    print('Error')

nums = []
for i in range(1, len(dat) + 1, 1):
    nums.append(i)
```

```python
rec = list()
for x in dat:
    rec.append(x['Cases'])
print(len(rec))
```

```python
plt.plot(rec)
plt.show()
```

```python
rec2 = list()
for x in dat:
    rec2.append(x['Recovered'])
print(len(rec2))
```

```python
plt.plot(nums, rec2)
plt.show()
```

```python
plt.plot(nums, rec, nums, rec2)
plt.show()
```

```python
# print(dat[-10]['Vac Given'])
numsvac = []
vac = []
count = 0
for x in dat:
    if x['Vac Given'] == None :
        pass
    else :
        count += 1
        vac.append(x['Vac Given'])
        numsvac.append(count)
```

```python
plt.plot(numsvac, vac)
plt.show()
```
